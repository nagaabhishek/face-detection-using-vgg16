```python
import cv2
from PIL import Image
from keras.models import load_model
import numpy as np
from keras.preprocessing import image
model=load_model('facefeatures_new_model.h5')

face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')

def face_extractor(img):
    faces = face_cascade.detectMultiScale(img, 1.3, 5)
    if faces is ():
        return None
    for (x, y, w, h) in faces:
        cv2.rectangle(frame, (x, y), (x+w, y+h), (255, 0, 0), 2)
        cropped_face=img[y:y+h,x:x+w]
    return cropped_face
        
video_capture = cv2.VideoCapture(0)
while True:
    _, frame = video_capture.read()
    face=face_extractor(frame)
    if type(face) is np.ndarray:
        face=cv2.resize(face,(224,224))
        im=Image.fromarray(face,'RGB')
        
        img_array=np.array(im)
        img_array=np.expand_dims(img_array,axis=0)
        pred=model.predict(img_array)
        print(pred)
        
        name=" None Matching"
        
        if(pred[0][0]>0.5):
            name='chetan'
            cv2.putText(frame,name,(50,50),cv2.FONT_HERSHEY_COMPLEX,1,(255,0,0),2)
        elif(pred[0][1]>0.5):
            name='daddy'
            cv2.putText(frame,name,(00,185),cv2.FONT_HERSHEY_SIMPLEX,1,(0,255,0),2)
        elif(pred[0][2]>0.5):
            name='abhishek'
            cv2.putText(frame,name,(00,285),cv2.FONT_HERSHEY_COMPLEX,1,(0,255,0),2)
        else:
            cv2.putText(frame,'not matching',(50,385),cv2.FONT_HERSHEY_COMPLEX,1,(255,0,0),2)
            
    else:
        cv2.putText(frame,'No Face Found',(50,50),cv2.FONT_HERSHEY_COMPLEX,1,(0,0,255),2)
    cv2.imshow('Video', frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
video_capture.release()
cv2.destroyAllWindows()
```
